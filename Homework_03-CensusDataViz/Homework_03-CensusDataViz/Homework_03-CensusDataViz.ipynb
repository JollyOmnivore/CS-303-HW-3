# python code to process, analyze, visualize trend #3


```python
data = pd.read_csv('adult.data', header=None, delimiter = ',')
df = pd.DataFrame(data.values, columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'])


i = (df['age'] >= 0) & (df['age'] < 18)
young = df['hours-per-week'][i]

i = (df['age'] >= 18) & (df['age'] < 65)
middle = df['hours-per-week'][i]

i = (df['age'] > 65) & (df['age'] < 100)
old = df['hours-per-week'][i]

plt.boxplot([young,middle,old])
plt.title('Hours worked by age')
plt.xlabel('Age')
plt.ylabel('Hours Worked')


plt.xticks([1,2,3],['<18','18 to 65','65+'] )



plt.show()
```

# plaintext explanation of trend #3. What did you find? How does your graph show this trend?

shows how people who are over the age of 18 to 65 work more more hours then younger and older people, it also shows a lot of old peopel still work instead of retire

```python
data = pd.read_csv('adult.data', header=None, delimiter = ',')
df = pd.DataFrame(data.values, columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'])




i = (df['occupation'] == ' Machine-op-inspct') 
oc1 = df['capital-gain'][i]

i = (df['occupation'] == ' Tech-support')
oc2 = df['capital-gain'][i]




i = (df['occupation'] == ' Adm-clerical')
oc3 = df['capital-gain'][i]


i = (df['occupation'] == ' Exec-managerial')
oc4 = df['capital-gain'][i]

plt.boxplot([oc1,oc2,oc3,oc4])

plt.title('capital gain based on occupation')
plt.xlabel('Occupation')
plt.ylabel('capital gain')


plt.xticks([1,2,3,4],['Machine-op-inspct','Tech-support','Adm-clerical','Exec-managerial'] )



plt.show()
```
